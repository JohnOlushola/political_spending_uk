```python
from transformers import pipeline

classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

```python
 labels = ["artifacts", "animals", "food", "birds",'Transport','advertising']
 hypothesis_template = 'This text is about {}.'
 sequence = "11 GE2019 - miner endorsement"
```

```python
prediction = classifier(sequence, labels, hypothesis_template=hypothesis_template, multi_class=True)

print(prediction['labels'][0])
```
